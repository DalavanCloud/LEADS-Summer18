# Comparing annotations from multiple platforms 

A selection of the following document was annotated by a human using two different annotation platforms (Recogito and brat), and then by the Mordecai full-text geoparser: https://www.biodiversitylibrary.org/item/19764#page/7/mode/1up
Place names found using each method are visualized below.

## Visualize place names extracted with human annotations in Recogito

```python
import os
import folium
from IPython.display import HTML
import pandas as pd

#Reads in data as data frame
data = os.path.join('data', '/Users/gretchenstahlman/Documents/Geoparsing/1meahgpqshvpt0.csv')
maptest_df = pd.read_csv(data)

#Drops missing values
maptest_df=maptest_df.dropna(subset=['longitude'])

maptest_df=maptest_df.dropna(subset=['latitude'])

#Creates list with coordinates
locations = maptest_df[['latitude', 'longitude']]
locationlist = locations.values.tolist()
len(locationlist)

#Defines map appearance and parameters
m = folium.Map(
    location=[-10, -75.25],
    zoom_start=4,
    tiles='Stamen Terrain'
)

#Adds markers to locations
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point]).add_to(m)
    
m
#m.save('/.../map1.html')
```
